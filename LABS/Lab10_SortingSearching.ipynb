```python
# Lab-10: Sorting and Searching Algorithms

# 1. Bubble Sort using Classes and Methods
class BubbleSort:
    def __init__(self, arr):
        self.arr = arr

    def sort(self):
        n = len(self.arr)
        for i in range(n):
            for j in range(0, n - i - 1):
                if self.arr[j] > self.arr[j + 1]:
                    self.arr[j], self.arr[j + 1] = self.arr[j + 1], self.arr[j]
        return self.arr

# 2. Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# 3. Selection Sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# 4. Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1
        
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
        
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

# 5. Linear Search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# 6. Binary Search (Array must be sorted)
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# 7. Binary Tree
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(self.root, value)

    def _insert(self, current, value):
        if value < current.value:
            if current.left is None:
                current.left = Node(value)
            else:
                self._insert(current.left, value)
        else:
            if current.right is None:
                current.right = Node(value)
            else:
                self._insert(current.right, value)

    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, current, result):
        if current:
            self._inorder(current.left, result)
            result.append(current.value)
            self._inorder(current.right, result)

# Demonstrations
print("1. Bubble Sort:")
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort = BubbleSort(arr.copy())
print("Sorted Array:", bubble_sort.sort())

print("\n2. Insertion Sort:")
arr = [64, 34, 25, 12, 22, 11, 90]
print("Sorted Array:", insertion_sort(arr.copy()))

print("\n3. Selection Sort:")
arr = [64, 34, 25, 12, 22, 11, 90]
print("Sorted Array:", selection_sort(arr.copy()))

print("\n4. Merge Sort:")
arr = [64, 34, 25, 12, 22, 11, 90]
print("Sorted Array:", merge_sort(arr.copy()))

print("\n5. Linear Search:")
arr = [10, 20, 30, 40, 50]
target = 30
print(f"Element {target} found at index:", linear_search(arr, target))

print("\n6. Binary Search:")
arr = [10, 20, 30, 40, 50]
target = 40
print(f"Element {target} found at index:", binary_search(arr, target))

print("\n7. Binary Tree:")
bt = BinaryTree()
elements = [50, 30, 20, 40, 70, 60, 80]
for el in elements:
    bt.insert(el)
print("Inorder Traversal:", bt.inorder())
```
